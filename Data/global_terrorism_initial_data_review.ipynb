```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
%matplotlib inline
```

```python
df = pd.read_csv('globalterrorismdb_0617dist.csv',encoding='latin1')
```

```python
df.head()
```

```python
df.shape
```

```python
events_by_year = df.groupby('iyear').count()
events_by_country = df.groupby('country_txt').count()
events_by_region = df.groupby('region_txt').count()
events_by_month = df.groupby('imonth').count()
events_by_city = df.groupby('city').count()
events_by_type = df.groupby('attacktype1_txt').count()
events_by_target = df.groupby('targtype1_txt').count()
```

```python
events_by_year = events_by_year[['eventid']]
events_by_year.plot(kind='bar',figsize=(15,10))
```

```python
events_by_country = events_by_country[['eventid']]
events_by_country.sort_values(by='eventid',ascending=False,inplace=True)
events_by_country = events_by_country.head(50)
```

```python
events_by_country.plot(kind='bar', figsize=(15,10))
```

```python
events_by_region = events_by_region[['eventid']].sort_values(by='eventid', ascending='true')
events_by_region.plot(kind='bar', figsize=(15,10))
```

```python
events_by_month = events_by_month[['eventid']]
events_by_month.plot(kind='bar',figsize=(15,10))
```

```python
events_by_city = events_by_city[['eventid']]
events_by_city.sort_values(by='eventid',inplace=True, ascending=False)
events_by_city = events_by_city.head(30)
events_by_city.plot(kind='bar', figsize=(15,10))
```

```python
events_by_type = events_by_type[['eventid']].sort_values(by='eventid',ascending=False)
events_by_type.plot(kind='bar', figsize=(15,10))
```

```python
events_by_target = events_by_target[['eventid']].sort_values(by='eventid', ascending=False)
events_by_target.plot(kind='bar', figsize=(15,10))
```
